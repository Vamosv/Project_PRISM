```python
# =============================================================================
# Personality Profile Module (with Agenda)
# =============================================================================
class PersonalityProfile:
    def __init__(self,
                 lm_familiarity: str = "Unknown",
                 lm_frequency_use: str = "Unknown",
                 age: str = "Unknown",
                 gender: str = "Unknown",
                 employment_status: str = "Unknown",
                 education: str = "Unknown",
                 marital_status: str = "Unknown",
                 english_proficiency: str = "Unknown",
                 study_locale: str = "Unknown",
                 religion: str = "Unknown",
                 ethnicity: str = "Unknown",
                 location: str = "Unknown",
                 agenda: str = "No specific agenda"):
        self.lm_familiarity = lm_familiarity
        self.lm_frequency_use = lm_frequency_use
        self.age = age
        self.gender = gender
        self.employment_status = employment_status
        self.education = education
        self.marital_status = marital_status
        self.english_proficiency = english_proficiency
        self.study_locale = study_locale
        self.religion = religion
        self.ethnicity = ethnicity
        self.location = location
        self.agenda = agenda

    def to_dict(self) -> dict:
        return {
            "lm_familiarity": self.lm_familiarity,
            "lm_frequency_use": self.lm_frequency_use,
            "age": self.age,
            "gender": self.gender,
            "employment_status": self.employment_status,
            "education": self.education,
            "marital_status": self.marital_status,
            "english_proficiency": self.english_proficiency,
            "study_locale": self.study_locale,
            "religion": self.religion,
            "ethnicity": self.ethnicity,
            "location": self.location,
            "agenda": self.agenda,
        }

    def __repr__(self):
        return (f"PersonalityProfile(lm_familiarity='{self.lm_familiarity}', "
                f"lm_frequency_use='{self.lm_frequency_use}', age='{self.age}', "
                f"gender='{self.gender}', employment_status='{self.employment_status}', "
                f"education='{self.education}', marital_status='{self.marital_status}', "
                f"english_proficiency='{self.english_proficiency}', study_locale='{self.study_locale}', "
                f"religion='{self.religion}', ethnicity='{self.ethnicity}', "
                f"location='{self.location}', agenda='{self.agenda}')")

# =============================================================================
# Conversation Turn and Log Module
# =============================================================================
class ConversationTurn:
    def __init__(self,
                 turn: int,
                 role: str,   # e.g. "model" (all agent responses) or "user" (for synthetic opening context, if needed)
                 content: str,
                 model_provider: str = None,
                 model_name: str = None,
                 score: int = None,
                 within_turn_id: int = None):
        self.turn = turn
        self.role = role
        self.content = content
        self.model_provider = model_provider
        self.model_name = model_name
        self.score = score
        self.within_turn_id = within_turn_id

    def to_dict(self) -> dict:
        return {
            "turn": self.turn,
            "role": self.role,
            "content": self.content,
            "model_provider": self.model_provider,
            "model_name": self.model_name,
            "score": self.score,
            "within_turn_id": self.within_turn_id
        }

class Conversation:
    def __init__(self, user_id: str, opening_prompt: str):
        self.conversation_id = f"c{uuid.uuid4().hex[:4]}"
        self.user_id = user_id
        self.opening_prompt = opening_prompt
        # We do not add the opening prompt as its own turn.
        self.conversation_history: List[ConversationTurn] = []

    def add_turn(self,
                 turn: int,
                 role: str,
                 content: str,
                 model_provider: str = None,
                 model_name: str = None,
                 score: int = None,
                 within_turn_id: int = None):
        ct = ConversationTurn(turn, role, content,
                              model_provider=model_provider,
                              model_name=model_name,
                              score=score,
                              within_turn_id=within_turn_id)
        self.conversation_history.append(ct)

    def export_history(self) -> List[dict]:
        return [turn.to_dict() for turn in self.conversation_history]

    def __repr__(self):
        return (f"Conversation(conversation_id='{self.conversation_id}', user_id='{self.user_id}', "
                f"opening_prompt='{self.opening_prompt}', "
                f"conversation_history={self.export_history()})")

# =============================================================================
# LM Agent Module Using Groq for Chat Completions
# =============================================================================
class LM_Agent:
    def __init__(self, name: str, personality: PersonalityProfile, provider: str, model: str):
        """
        model: The Groq model identifier, e.g. "meta-llama/llama-4-scout-17b-16e-instruct"
        """
        self.name = name
        self.personality = personality
        self.provider = provider
        self.model = model

    def generate_response(self, conversation_context: List[ConversationTurn], topic: str = None) -> dict:
        """
        Build the prompt from the conversation context (which may include the opening prompt if no turns exist)
        and the agentâ€™s personality & agenda. Call Groq to generate a response.
        Returns a dictionary with response details.
        """
        prompt = self._build_prompt(conversation_context, topic)
        chat_completion = client.chat.completions.create(
            messages=[{"role": "user", "content": prompt}],
            model=self.model,
        )
        content = chat_completion.choices[0].message.content
        return {"content": content, "score": 100, "within_turn_id": 0}

    def _build_prompt(self, conversation_context: List[ConversationTurn], topic: str = None) -> str:
        # If context is empty (i.e. in round 0), include the opening prompt as a synthetic user message.
        if not conversation_context:
            context_text = f"User: {orchestrator.conversation.opening_prompt}"
        else:
            context_text = "\n".join(
                [f"{turn.role.capitalize()}: {turn.content}" for turn in conversation_context]
            )

        base_prompt = f"{context_text}\n"
        if topic:
            base_prompt += f"\nRandom Topic: {topic}\n"
        base_prompt += f"\n[Respond with personality: {self.personality} | Agenda: {self.personality.agenda}]"
        return base_prompt

    def __repr__(self):
        return (f"LM_Agent(name='{self.name}', provider='{self.provider}', "
                f"model='{self.model}', personality={self.personality})")

# =============================================================================
# Conversation Orchestrator Module
# =============================================================================
class ConversationOrchestrator:
    def __init__(self, user_id: str, opening_prompt: str, lm_agents: List[LM_Agent]):
        self.conversation = Conversation(user_id, opening_prompt)
        self.lm_agents = lm_agents
        # Turn counter: each round (shared by all agent messages) is one turn.
        self.current_turn = 0
        # Store all agentsâ€™ personality profiles.
        self.personality_profiles = {
            agent.name: agent.personality.to_dict() for agent in lm_agents
        }

    def get_random_topic(self) -> str:
        topics = [
            "Discuss the impact of AI on society.",
            "What are the challenges of remote work?",
            "How do cultural differences shape communication?",
            "Share thoughts on future technological trends."
        ]
        return random.choice(topics)

    def get_context_for_round(self) -> List[ConversationTurn]:
        """
        For round 0, the context is empty (so that the agent uses the opening prompt stored in the conversation).
        For later rounds, use the accumulated conversation history.
        """
        if self.current_turn == 0:
            return []  # No conversation turns yet; agent's prompt-building will pick up the opening prompt.
        else:
            return self.conversation.conversation_history

    def next_round(self):
        """
        In each round, every agent produces one message.
        All messages in a round share the same turn number.
        """
        topic = self.get_random_topic()  # Optional: add an extra topic for flavor.
        context = self.get_context_for_round()
        for agent in self.lm_agents:
            response = agent.generate_response(context, topic)
            self.conversation.add_turn(
                turn=self.current_turn,
                role="model",
                content=response["content"],
                model_provider=agent.provider,
                model_name=agent.name,
                score=response["score"],
                within_turn_id=response["within_turn_id"]
            )
        self.current_turn += 1

    def export_conversation(self) -> dict:
        return {
            "conversation_id": self.conversation.conversation_id,
            "user_id": self.conversation.user_id,
            "opening_prompt": self.conversation.opening_prompt,
            "personality_profiles": self.personality_profiles,
            "conversation_history": self.conversation.export_history()
        }

    def run_conversation(self, max_rounds: int = 5):
        for _ in range(max_rounds):
            self.next_round()


# =============================================================================
# Main Application - Putting It All Together
# =============================================================================
def main():
    # Create a personality profile for an agent with a persona.
    persona_with_agenda = PersonalityProfile(
        lm_familiarity="Somewhat familiar",
        lm_frequency_use="Once per month",
        age="25-34 years old",
        gender="Female",
        employment_status="Working full-time",
        education="Vocational",
        marital_status="Never been married",
        english_proficiency="Native speaker",
        study_locale="canada",
        religion="No Affiliation",
        ethnicity="White",
        location="Northern America",
        agenda="I want to inspire creativity in poetry."
    )
    
    # Create a default personality profile for an agent with no specific persona.
    default_persona = PersonalityProfile(
        agenda="I am here to assist without bias."
    )

    # Specify the Groq model name.
    model_name = "meta-llama/llama-4-scout-17b-16e-instruct"

    # Create two LM agents.
    lm_agent1 = LM_Agent(name="LM-Agent-Persona", personality=persona_with_agenda, provider="groq", model=model_name)
    lm_agent2 = LM_Agent(name="LM-Agent-Default", personality=default_persona, provider="groq", model=model_name)
    agents = [lm_agent1, lm_agent2]

    # Define the opening prompt (which is not stored as its own turn but used as context).
    opening_prompt = "I would like to write a poem about my white cat."
    user_id = "user9"

    # Initialize the conversation orchestrator.
    global orchestrator  # Allow access to orchestrator in LM_Agent for the opening prompt
    orchestrator = ConversationOrchestrator(user_id, opening_prompt, agents)

    # Run the conversation for a fixed number of rounds.
    orchestrator.run_conversation(max_rounds=3)
    
    # Export the conversation log (including personality profiles) as JSON.
    conversation_data = orchestrator.export_conversation()
    
    # Dump the conversation data to a JSON file.
    with open("conversation_dump.json", "w") as f:
        json.dump(conversation_data, f, indent=2)
    
    # Also print the JSON output to the console.
    print("Final Conversation Log:")
    print(json.dumps(conversation_data, indent=2))

if __name__ == "__main__":
    main()
```
